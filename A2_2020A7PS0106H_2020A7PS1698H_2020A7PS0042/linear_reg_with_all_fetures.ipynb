```python
import random
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline 
import warnings
warnings.filterwarnings("ignore")
```

```python
# load the dataset
df = pd.read_csv('FODS-A2.csv')
df.head()
```

```python
df.isnull().sum()#any missing values in column
```

```python
df[df.isnull().any(axis=1)]# any missing values in row
X=df.iloc[:,0:26]
y=df.iloc[:,-1]
#min max normalize
X_max=X.max()
X_min=X.min()
X=(X-X_min)/(X_max-X_min)
y_max=y.max()
y_min=y.min()
y=(y-y_min)/(y_max-y_min)
# train test split

```

```python
order = list(range(len(y)))

shuffle = random.sample(order, int(len(order) * 0.8))

rem = [i for i in order if i not in shuffle]
```

```python
X_train = X.iloc[shuffle,:]
y_train = y.iloc[shuffle]
X_test = X.iloc[rem,:]
y_test = y.iloc[rem]
X_train
```

```python
def find_theta(x, y):
    theta = np.dot(np.linalg.pinv(np.dot(x.T, x)), np.dot(x.T, y))
    return theta
```

```python
def predict(x, theta):
    pred = np.dot(x, theta)
    return pred
```

```python
def linear_regression(train_x,train_y,test_x,test_y):
    theta = find_theta(train_x,train_y)
    y_pred_train = predict(train_x, theta)
    # calculate rmse train
    y_pred_train = np.array(y_pred_train)
    train_y = np.array(train_y)
    rmse_train = np.sqrt(np.sum((y_pred_train - train_y)**2)/len(train_y))

    # calculate rmse test
    y_pred_test = predict(test_x, theta)
    y_pred_test = np.array(y_pred_test)
    rmse_test = np.sqrt(np.sum((y_pred_test - test_y)**2)/len(train_y))*2

    return rmse_train,rmse_test
```

```python
X1 = np.ones((len(X_train), 1))
X2 = np.ones((len(X_test), 1))
X_train_selected = np.c_[(X1, X_train)]
X_test_selected = np.c_[X2, X_test]
rmse_train,rmse_test=linear_regression(X_train_selected,y_train,X_test_selected,y_test)
```

```python
print("Training Error With all features =",rmse_train)
print("Testing Error With all features =",rmse_test)
```
