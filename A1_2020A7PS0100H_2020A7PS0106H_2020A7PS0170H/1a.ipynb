```python
import numpy as np
from scipy.stats import beta
import matplotlib.pyplot as plt

#Enter the parameters:

a=2
b=2

# Beta distribution

x = np.linspace(0, 1, 100)
y = beta.pdf(x, a, b)
plt.plot(x, y)
plt.title("Beta Distribution of s",fontsize='15')
plt.xlabel("s",fontsize='15')
plt.ylabel("y-Probability",fontsize='15')
plt.show()
```
