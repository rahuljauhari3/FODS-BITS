```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df=pd.read_csv("fods_1.csv")
```

```python
def normalize(X):
    mean = np.mean(X,axis=0)
    sd = np.std(X,axis=0)
    X = (X-mean)/sd
    return X
```

```python
X=normalize(np.array(df.iloc[:,[0,2]])).T
Y=normalize(np.array(df.iloc[:,[2]])).T
m=X.shape[1]
Y.shape
```

```python
def unregularized_cost(w1,w2):
    m = X.shape[1]
    ans=[]
    for j in range(w1.shape[0]):
        ans.append(np.average((w1[j]*X[0,:]+w2[j]*X[1,:]-Y[0,:])**2)/2)
    return np.array(ans).reshape(len(ans),1)
```

```python
W=np.zeros((2,1))
ranges=np.arange(-3,3,0.05)
W1,W2=np.meshgrid(ranges.flatten(),ranges.flatten())
Z = unregularized_cost(W1.flatten(), W2.flatten()).reshape(W1.shape[0],W1.shape[1])
```

```python
plt.figure(figsize=(8,8))
plt.contour(W1, W2, Z, colors='gray',levels=200)
plt.title("Contour for q = 0.5")
plt.xlabel("W1")
plt.ylabel("W2")
```
